```python
# import scipy.linalg as lin
import pickle
import numpy as np
import os
# from linalg_helpers import compute_powers

import matplotlib.pyplot as plt
```

```python
valA = "ortho_haar"

if "ortho" in valA:
    valC = "_ident_C"

path = f"/data/shared/ICL_Kalman_Experiments/train_and_test_data/{valA}/train_{valA}{valC}_state_dim_5"

with open(f"{path}.pkl", "rb") as f:
    ys = pickle.load(f)

sys = 2
print(f"len(ys) = {len(ys)}")
print(f"len(ys[sys]['obs']) = {len(ys[sys]['obs'])}")
```

```python
num_sys = 3
end_ind = 250
fig, ax = plt.subplots(num_sys, 1, figsize=(10, num_sys*5))
for sys in range(num_sys):
    y = ys[sys]['obs']
    print(f"y.shape = {y.shape}")
    for i in range(y.shape[1]):
        ax[sys].scatter(np.arange(end_ind), y[:end_ind, i], label=f"y[:, {i}]")

    ax[sys].set_title(f"System {sys}")
    ax[sys].set_xlabel("Index")
    ax[sys].set_ylabel("Value")
    ax[sys].legend()
    #set minor gridlines on every 1 x-axis tick
    ax[sys].xaxis.set_minor_locator(plt.MultipleLocator(1))
    #set major gridlines on every 10 x-axis ticks
    ax[sys].xaxis.set_major_locator(plt.MultipleLocator(10))

    ax[sys].grid(which='major', linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.show()

fig_path = f"{path}_end_ind_{end_ind}.pdf"

fig.savefig(fig_path, bbox_inches='tight', format='pdf')
```
